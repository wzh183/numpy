```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize'] = (5,3)
```

```python
fig = plt.figure('f1',figsize=(6,4),dpi=60)
plt.plot()
```

```python
x = np.arange(0,51)
y = x**20
# 创建图形对象 设置参数
plt.figure('f1',figsize=(4,2),dpi=100,facecolor='pink')
plt.plot(x,y)
```

```python
import math
x=list(np.arange(-40,40,0.1))
y=[]
for i in range(len(x)):
    y.append(math.sin(1/x[i]))
plt.figure('f1',figsize=(4,2),dpi=100,facecolor='pink')
plt.plot(x,y)
```

```python
fig = plt.figure(figsize=(4,2),facecolor='pink')
# 从画布起始位置绘制,宽高喝画布相同
axes1 = fig.add_axes([0,0,1,1])
print(axes1)
# 画布20%开始绘制,宽高是画布的50%
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
print(axes2)
```
