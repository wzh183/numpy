```python
from matplotlib import pylab as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize'] = (5,3)
```

```python
countries = ['挪威','德国', '中国','美国','瑞典']
# 金牌个数
gold_medal = [16,12,9,8,8]
silver_medal = [8,10,4,10,5]
bronze_medal = [13,5,2,7,5]
width = 0.2
all_medal = [37,27,15,25,18]
# 将x轴转化为数值
x = np.arange(len(countries))
# 设置画布
fig = plt.figure(figsize=(6,4),dpi=150)
# 将画布分为两行三列

ax1 = fig.add_subplot(221)
ax1.set_title('奖牌榜')
ax1.tick_params(axis='x',rotation=45)
ax1.bar(countries, all_medal,color='r')

ax2 = fig.add_subplot(222)
ax2.set_title('金牌榜')
ax2.tick_params(axis='x',rotation=45)
ax2.bar(countries, gold_medal,color='gold')

ax2 = fig.add_subplot(223)
ax2.set_title('银牌榜')
ax2.tick_params(axis='x',rotation=45)
ax2.bar(countries, silver_medal,color='silver')

ax2 = fig.add_subplot(224)
ax2.set_title('铜牌榜')
ax2.tick_params(axis='x',rotation=45)
ax2.bar(countries, bronze_medal,color='brown')
plt.tight_layout()
```

```python
countries = ['挪威','德国', '中国','美国','瑞典']
# 金牌个数
gold_medal = np.array([16,12,9,8,8])
silver_medal = np.array([8,10,4,10,5])
bronze_medal = np.array([13,5,2,7,5])
# 绘制堆叠图
width=0.3
# 绘制金牌
plt.bar(countries,gold_medal,color='gold',label='金牌',bottom=silver_medal+bronze_medal,width=width)
plt.bar(countries, silver_medal, color='silver', label='银牌', bottom=bronze_medal, width=width)
plt.bar(countries,gold_medal,color='brown',label='铜牌',width=width)
plt.ylabel('奖牌数')
plt.legend(loc='upper right')
for i in range(len(countries)):
    max_y = gold_medal[i]+silver_medal[i]+bronze_medal[i]
    plt.text(countries[i], max_y, max_y, va='bottom', ha='center')
```

```python
movie = ['千与千寻','玩具总动员','黑衣人:全球通缉']
real_day1=[4053,7548,6543]
real_day2=[1840,4013,3421]
real_day3=[2080,1673,2342]
width = 0.2
# 将x轴转化为数值
x = np.arange(len(movie))
# 确定x轴的起始位置
r1_x = x
r2_x = x + width
r3_x = x + 2 * width
# 绘制
plt.bar(r1_x,real_day1,color='gold',width=width,label='千与千寻')
plt.bar(r2_x,real_day2,color='silver',width=width,label='玩具总动员')
plt.bar(r3_x,real_day2,color='brown',width=width,label='黑衣人:全球通缉')
# 把x轴转化为国家
plt.xticks(x+width,labels=movie)
plt.legend(loc='upper right')
for i in range(len(movie)):
    plt.text(r1_x[i],real_day1[i],real_day1[i],va='bottom',ha='center',fontsize=8)
    plt.text(r2_x[i],real_day2[i],real_day2[i],va='bottom',ha='center',fontsize=8)
    plt.text(r3_x[i],real_day3[i],real_day3[i],va='bottom',ha='center',fontsize=8)
```

```python
movie = ['千与千寻','玩具总动员','黑衣人:全球通缉']
real_day1=np.array([4053,7548,6543])
real_day2=np.array([1840,4013,3421])
real_day3=np.array([2080,1673,2342])
width = 0.2
# 将x轴转化为数值
x = np.arange(len(movie))
plt.bar(movie, real_day1, color='gold', label='千与千寻', bottom=real_day2+real_day3, width=width)
plt.bar(movie, real_day2, color='silver', label='玩具总动员', bottom=real_day3, width=width)
plt.bar(movie, real_day3, color='brown', label='黑衣人:全球通缉', width=width)
plt.ylabel('票房数')
plt.legend(loc='upper right')
for i in range(len(movie)):
    max_y = real_day1[i]+real_day2[i]+real_day3[i]
    plt.text(movie[i], max_y, max_y, va='bottom', ha='center')
```
