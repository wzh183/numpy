```python
from matplotlib import pylab as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize'] = (5,3)
```

```python
# x 轴数据
x = range(5)
# y 轴数据
data = [5,20,15,25,10]
# 设置标题
plt.title('基本柱状图')
# 网格
plt.grid(ls='--',alpha=0.5)

# 绘制柱状图
plt.bar(x,data)
```

```python
# x 轴数据
x = range(5)
# y 轴数据
data = [5,20,15,25,10]
# 设置标题
plt.title('基本柱状图')
# 网格
plt.grid(ls='--',alpha=0.5)

# 绘制柱状图 bottom 形状要一致,每个柱状图都要给
# plt.bar(x,data,bottom=[10,20,5,0,10])

# plt.bar(x,data,color=['yellow','black'])  # 多个
plt.bar(x,data,facecolor='pink')   # 单个
```

```python
# 描边
plt.bar(x,data,ec='r',ls='--',lw='1')
```

```python
countries = ['挪威','德国', '中国','美国','瑞典']
# 金牌个数
gold_medal = [16,12,9,8,8]
silver_medal = [8,10,4,10,5]
bronze_medal = [13,5,2,7,5]
width = 0.2

# 将x轴转化为数值
x = np.arange(len(countries))
# 确定x轴的起始位置
gold_x = x
silver_x = x + width
bronze_x = x + 2 * width
# 绘制
plt.bar(gold_x,gold_medal,color='gold',width=width,label='金牌')
plt.bar(silver_x,silver_medal,color='silver',width=width,label='银牌')
plt.bar(bronze_x,bronze_medal,color='brown',width=width,label='金牌')
# 把x轴转化为国家
plt.xticks(x+width,labels=countries)
plt.legend(loc='upper right')
for i in range(len(countries)):
    plt.text(gold_x[i],gold_medal[i],gold_medal[i],va='bottom',ha='center',fontsize=8)
    plt.text(silver_x[i],silver_medal[i],silver_medal[i],va='bottom',ha='center',fontsize=8)
    plt.text(bronze_x[i],bronze_medal[i],bronze_medal[i],va='bottom',ha='center',fontsize=8)
```
