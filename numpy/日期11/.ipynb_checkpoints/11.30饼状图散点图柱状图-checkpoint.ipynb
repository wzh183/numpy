```python
from matplotlib import pylab as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize'] = (5,5)
```

```python
# 指定分组个数
bins = 10
fig,ax = plt.subplots()
# 分别生成10000, 5000, 2000,个数字
x_values = [np.random.randn(n) for n in [10000,5000,2000]]
ax.hist(x_values,bins=bins,label=list('123'))
ax.legend()
```

```python
x1 = np.random.randint(140,180,200)
x2 = np.random.randint(140,180,200)
plt.hist([x1,x2],bins=10,stacked=True,edgecolor='white',label='12')
plt.legend()
```

```python
# 定义标签
label = ['娱乐','育儿','饮食','房贷','交通','其他']
# 定义每个标签所占数量
x = [200,500,1200,7000,200,900]
plt.title('家庭支出')
# 绘图
plt.pie(x,labels=label,autopct='%.2f%%', pctdistance=1,labeldistance=1.5)
plt.legend()
```

```python
# 定义标签
label = ['娱乐','育儿','饮食','房贷','交通','其他']
# 定义每个标签所占数量
x = [200,500,1200,7000,200,900]
plt.title('家庭支出')
explode = (0.03,0.05,0.06,0.04,0.08,0.21)

# 绘图
plt.pie(x,labels=label,autopct='%.2f%%',explode=explode, pctdistance=1,labeldistance=1.5)
plt.legend()
```

```python
# x轴数据
x = np.array([1,2,3,4,5,6,7,8])
# y轴的数据
y = x **3

# 绘图
plt.scatter(x,y)
```

```python
# x轴数据
x = np.array([1,2,3,4,5,6,7,8])
# y轴的数据
y = x **3
s = (20 * np.random.randn(8)**2)
plt.scatter(x,y,s=s)
```

```python
# x轴数据
x = np.random.rand(50)
# y轴的数据
y = np.random.rand(50)
s = (10 * np.random.randn(50)**2)
clors = np.random.rand(50)
# 绘图
plt.scatter(x,y,s=s,c=clors)
```
