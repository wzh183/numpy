```python
from matplotlib import pylab as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize'] = (5,3)
```

```python
fig = plt.figure(figsize=(4,2), facecolor='g')
x = np.arange(0,50,2)
y = x ** 2
# 区域1
axese1 = fig.add_axes([0.0,0.0,1.0,1.0])
axese1.plot(x,y)
axese2 = fig.add_axes([0.7,0.7,0.3,0.3])
axese2.plot(x,y)
```

```python
fig = plt.figure(figsize=(4,2),facecolor='g')
x = np.arange(0,50,2)
y = x ** 3
axes1 = fig.add_axes([0.0,0.0,1.0,1.0])
axes1.set_title('axes1')
axes1.set_xlabel('X axes1')
index = [x [i] for i in range(len(x))]
# print(index)
axes1.set_xticks(index)
axes1.plot(x,y)
axes2 = fig.add_axes([0.6,0.6,0.4,0.4])
axes2.set_title('axes2')
axes2.set_xlabel('X axes2')
axes2.plot(x,y)
```

```python
fig = plt.figure(figsize=(4,2),facecolor='g')
x = np.arange(0,50,2)
y = x ** 3
axes1 = fig.add_axes([0.0,0.0,1.0,1.0])
axes1.set_title('axes1')
axes1.set_xlabel('X axes1')
index = [x [i] for i in range(len(x))]
# print(index)
axes1.set_xticks(index)
axes1.plot(x,y,label='线条1')

axes1.plot(x,y)
axes1.legend()
y1 = x * 2
axes2 = fig.add_axes([0.2,0.2,0.4,0.4])
axes2.set_title('axes2')
axes2.set_xlabel('X axes2')
axes2.plot(x,y1,label='线条2')
axes2.plot(x,y1)
axes2.legend()
```

```python
fig,axes = plt.subplots(2,2)
# 第一个绘图区域
ax1 = axes[0][0]
x = np.arange(1,5)
y = x ** 2
ax1.plot(x,y)
ax1.set_title('平方')
# 第二个绘图区域
ax2 = axes[0][1]
ax2.plot(x,np.sqrt(x))
ax2.set_title('平方根')
# 第三个绘图区域
ax3 = axes[1][0]
ax3.plot(x,np.exp(x))
ax3.set_title('指数')
# 第四个绘图区域
ax4 = axes[1][1]
ax4.plot(x,np.log10(x))
ax4.set_title('指数')
plt.tight_layout()
```

```python
a, b = plt.subplots(1,2)
# 设置画布的大小和分辨率
a.set_figheight(3)  # 73*3
a.set_figwidth(8)   # 73 *8
a.set_facecolor('pink')
# 画图区域1----------------------
x = np.arange(-50,50)
y = x**2
b[0].plot(x,y)
# 画图区域2----------------------
b[1].spines['top'].set_color('none')
b[1].spines['right'].set_color('none')
b[1].spines['left'].set_position(('axes',0.5))
b[1].spines['bottom'].set_position(('axes',0.0))
b[1].plot(x,y)

```
