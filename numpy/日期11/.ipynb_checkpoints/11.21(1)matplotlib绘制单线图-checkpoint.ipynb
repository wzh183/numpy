```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(-50,51)
y = x**2
print(x)
print(y)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-100, 100)
y = x + 1
# 设置图标名称
# plt.title("y=x+1")
plt.title("第一个一次函数")  # 会报错,不支持中文

plt.plot(x, y)
```

```python
# 配置中文编码
plt.rcParams['font.sans-serif']=['fangsong']
x = np.arange(-100, 100)
y = x + 1
# 设置图标名称
plt.title("第1个1次函数")

plt.plot(x, y)
```

```python
# 设置中文之后,负数符号不显示了
# 配置中文编码
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['fangsong']
x = np.arange(-100, 100)
y = x + 1
# 设置图标名称
plt.title("第1个1次函数")
plt.plot(x, y)
```

```python
# 中文显示
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['fangsong']
# 创建x为-10 到10 的整数
x = np.arange(-10,11)
y = x ** 2
# 设置标题
plt.title('y=x**2, x的取值范围是[-10,10]')
# 设置x轴,y轴名称
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.plot(x,y)
```

```python
x = np.arange(5,15)
y = 1/x
# 设置中文标题
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['fangsong']
plt.title('y=1/x, x的取值范围是[5,14]', fontsize=12)
# 设置x轴,y轴名称
plt.xlabel('x轴')
plt.ylabel('y轴')
# 设置线条的粗细
plt.plot(x,y, linewidth=5)
plt.plot
```

```python
# 在一张图片中绘制多条线条
x = np.arange(-10, 11)
y1 = x**2
y2 = x+1
plt.plot(x,y1, linewidth=5)
plt.plot(x,y2, linewidth=5)
```

```python
# 日期
times = np.arange(1990, 2020).astype(np.str_)
# 销量
sales = np.random.randint(500,2000,size=len(times))
# 绘图
# plt.xticks(range(0,len(times),2)),['%s年'%i for i in times],rotation=90)
plt.xticks(range(0,len(times),2), rotation=45)
plt.plot(times,sales)                                  
```
