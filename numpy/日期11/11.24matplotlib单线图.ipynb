```python
x = np.linspace(-np.pi,np.pi,endpoint=True)
c = np.cos(x)
s = np.sin(x)
plt.plot(x,c,label='cos')
plt.plot(x,s,label='sin')
plt.legend()
plt.grid(True,linestyle="--", color="black", linewidth="0.5")
```

```python
# 创建x轴
x = np.arange(-50,51)
# 创建y轴
y = x ** 2
# 获取当前的坐标轴
ax = plt.gca()
# 通过坐标轴确定最top, botttom,left,right上,下,左,右
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# 移动下轴到指定位置 
# position位置有三种 data, outwart(向外), axes
# axes
ax.spines['left'].set_position(('axes',0.5)) # 最左边
# data 表示按数值移动,给的数值代表挪动到y轴的刻度
# ax.spines['bottom'].set_position(('data',0.0))
plt.ylim(0,y.max())  # 设置y轴的取值范围
# 绘图
plt.plot(x,y)
```

```python

import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d
a,b = np.mgrid[-2:0.5:20j,-2:0.5:20j]
#测试数据
c=a*np.exp(-a**2-b**2)
#三维图形
ax = plt.subplot(111, projection='3d')
ax.set_title('图形');
ax.plot_surface(a,b,c,rstride=2, cstride=1, cmap=plt.cm.Spectral)
#设置坐标轴标签
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('C')
```
