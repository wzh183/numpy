```python
x = np.linspace(-np.pi,np.pi,endpoint=True)
c = np.cos(x)
s = np.sin(x)
plt.plot(x,c,label='cos')
plt.plot(x,s,label='sin')
plt.legend()
plt.grid(True,linestyle="--", color="black", linewidth="0.5")
```

```python

import matplotlib.pyplot as plt
import numpy as np
 
 
barSlices = 12
 
 
theta = np.linspace(0.0, 2*np.pi, barSlices, endpoint=False)
r = 30*np.random.rand(barSlices)
 
 
plt.polar(theta, r, color="chartreuse", linewidth=2,
          marker="*", mfc="b", ms=10)

```

```python

import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d
a,b = np.mgrid[-2:0.5:20j,-2:0.5:20j]
#
c=a*np.exp(-a**2-b**2)
#
ax = plt.subplot(111, projection='3d')
ax.set_title('');
ax.plot_surface(a,b,c,rstride=2, cstride=1, cmap=plt.cm.Spectral)
#
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('C')
```
