```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# x 轴
times = ['2015/8/1','2015/10/12','2015/123','2016/3/4','2015/5/15','2016/7/26','2016/10/16','2016/12/17']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['fangsong']
plt.title('收入-支出图')
plt.xlabel('日期')
plt.ylabel('金额')
# 收入
income = np.random.randint(500,2000,size=len(times))
#支出
expense = np.random.randint(300,1500,size=len(times))
# 绘制图形
plt.xticks(range(len(times)),rotation=60)
plt.plot(times,income,label='收入')
plt.plot(times,expense,label='支出')
# 图例
plt.legend(loc='upper right')
```

```python
# x 轴
times = ['2015/8/1','2015/10/12','2015/123','2016/3/4','2015/5/15','2016/7/26','2016/10/16','2016/12/17']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['fangsong']
plt.title('收入-支出图')
plt.xlabel('日期')
plt.ylabel('金额')
# 收入
income = np.random.randint(500,2000,size=len(times))
#支出
expense = np.random.randint(300,1500,size=len(times))
# 绘制图形
plt.xticks(range(len(times)),rotation=60)
plt.plot(times,income,label='收入')
plt.plot(times,expense,label='支出')
# 图例
plt.legend(loc='upper right')
for a,b in zip(times,income):
    plt.text(a, b, '%s万'%b)
for a,b in zip(times,expense):
    plt.text(a, b, '%s万'%b)
```

```python
# x 轴
sales = ['109','150','172','260','273','333','347','393','402','446','466','481','499','504','513','563','815','900','930','961']
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['fangsong']
plt.title('销量年份图')
plt.xlabel('年份')
plt.ylabel('销量')
# 收入
year = np.random.randint(2000,2020,size=len(sales))
# 绘制图形
plt.xticks(range(len(year)),rotation=60)
plt.plot(sales,year,label='销量趋势')
plt.legend(loc='upper right')
```
