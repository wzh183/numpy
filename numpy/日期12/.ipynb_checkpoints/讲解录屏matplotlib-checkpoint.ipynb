```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif']=['SimHei']  # 设置中文显示
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['figure.dpi']=100
plt.rcParams['figure.figsize']=(5,3)
```

```python
知识点1：
绘制曲线图,要求如下:
•	1.显示图形名称
•	2.为x轴和y轴定义说明信息
•	3.使用中文显示
已知数据为2000年到2020年20年之间的销量情况.
•	sales = [109, 150, 172, 260, 273, 333, 347, 393, 402, 446, 466, 481, 499,504, 513, 563, 815, 900, 930, 961] 要求:
•	1.绘制年度销量线形图.x轴为年份,y轴为销量.
•	2.显示图例
```

```python
x = np.arange(2000,2020).astype(np.str_)
# print(x,len(x))
y = [109, 150, 172, 260, 273, 333, 347, 393, 402, 446, 466, 481, 499,504, 513, 563, 815, 900, 930, 961]
# print(y, len(y))
plt.title('销量图')
plt.xlabel('年份')
plt.ylabel('销量')
plt.xticks(range(len(x)),rotation=45,color='pink')
plt.plot(x,y,label='销量')
```

```python
知识点2：
国家奖牌数量变化
countries = ['挪威','德国','中国','美国','瑞典']
gold_medal = [16,12,9,8,8]
silver_medal = [8,10,4,10,5]
bronze_medal=[13,5,2,7,5]
作业,按照以上数据
- 1.绘制水平同位置多柱状图
- 2.绘制水平堆叠图
```

```python
countries = ['挪威','德国','中国','美国','瑞典']
# 奖牌的个数
gold_medal = [16,12,9,8,8]
silver_medal = [8,10,4,10,5]
bronze_medal=[13,5,2,7,5]
width = 0.2
x = np.arange(len(countries))
gold_x = x
silver_x = x + width
bronze_x = x + width * 2
# 绘制
plt.bar(gold_x,gold_medal,color='gold',width=width,label='金牌')
plt.bar(silver_x,silver_medal,color='silver',width=width,label='银牌')
plt.bar(bronze_x,bronze_medal,color='brown',width=width,label='铜牌')
# 把x轴转化为国家
plt.xticks(x+width,labels=countries)
plt.legend(loc='upper right')
```

```python
countries = ['挪威','德国','中国','美国','瑞典']
# 奖牌的个数
gold_medal = np.array([16,12,9,8,8])
silver_medal = np.array([8,10,4,10,5])
bronze_medal=np.array([13,5,2,7,5])
width = 0.3
plt.bar(countries,gold_medal,color='gold',label='金牌',bottom=silver_medal+bronze_medal,width=width)
plt.bar(countries,silver_medal,color='silver',label='银牌',bottom=bronze_medal,width=width)
plt.bar(countries,bronze_medal,color='brown', label='铜牌',width=width)
plt.ylabel('奖牌数')
plt.legend(loc='upper right')
```

```python
知识点3：
家庭消费支出
消费类别：娱乐、育儿、饮食、房贷、交通、其他
对应金额：200、500、1200、7000、200、900
按照以上数据
绘制饼状图并显示百分比
```

```python
label = ['娱乐','育儿','饮食','房贷','交通','其他']
x = [200, 500, 1200, 7000, 200, 900]
plt.title('家庭支出')
# 绘图
plt.pie(x, labels=label,autopct='%.2f%%',pctdistance=1,labeldistance=1.5)
plt.legend()
```
